# Battery discharge functional test report 

### Methodology: 

  * Real SMM+ WiFi transmission every 2 minutes. Around 30 seconds of BLE transmission and 15-20s WiFi 
transmission.

  * **30 TX events per hour (estimated)**

  * Voltage is measured at 50Hz sampling rate and stored offline

  * Data is sent to test API and total transmissions are counted.

  * Measuring setup is Raspberry PI 2B+ with INA219 breakout connected by i2c.

  * The measurement script is not production quality yet. Is expected to hang sometimes.

### For completness: 

  * Espressif WROOM-02 with AT firmware 1.6
  * SMM+ PCB v3
  * WiFi AP Ubiquiti Unifi PRO 

```
import pandas as pd
import numpy as np
import glob
from pandas import Series
from numpy import datetime64
import matplotlib.pyplot as plt
import datetime

from google.colab import drive
drive.mount('/content/gdrive/')
```

```
###############################################
##### DATASETS LOAD AND PREPROCESS ######
###############################################

## PKCELL
dtypes = [np.float64, np.int32, np.float64, np.datetime64,np.float64]
#pkcellFilesPath = glob.glob('./PKCELL/*.csv')
pkcellFilesPath = glob.glob('/content/gdrive/My Drive/PROJECTS/B2B/PIKOLIN/SMM+/Energy estimation/Battery tests/PKCELL/*.csv')
pkcell_dataframe = pd.DataFrame()



for file in pkcellFilesPath:
    print("Appending file"+file+" to dataframe")
    pkcell_dataframe = pkcell_dataframe.append(pd.read_csv(file,header=None,dtype={'Voltage':float,'WiFi TX count':int,'Last TX time': float, 'timestamp':np.datetime64, 'WiFi pin ON': int}))
pkcell_dataframe.columns=['Voltage','WiFi TX count','Last TX time','timestamp','WiFi pin ON'] 
pkcell_dataframe['timestamp']= pkcell_dataframe['timestamp'].astype(np.datetime64)


### Data cleaning, voltage under 1.5V means no battery connected.
print("Data cleaning")
pkcell_dataframe = pkcell_dataframe.drop(pkcell_dataframe[pkcell_dataframe.Voltage < 1.5].index)


### Time axis transformation, from timestamp to time delta 
print("Processing time axis")
timeStartPkcell = pkcell_dataframe.iloc[0,3]
timeAxisPkcell = ((pkcell_dataframe.iloc[:,3] - timeStartPkcell))
timeAxisPkcell = (timeAxisPkcell / np.timedelta64(1, 's')) *1/3600

   
print(pkcell_dataframe.describe())
print(pkcell_dataframe.info())

## MAXELL 

maxellAPiCsv = ('/content/gdrive/My Drive/PROJECTS/B2B/PIKOLIN/SMM+/Energy estimation/Battery tests/MAXELL/maxell_api.info')
maxellFilesPath = glob.glob('/content/gdrive/My Drive/PROJECTS/B2B/PIKOLIN/SMM+/Energy estimation/Battery tests/MAXELL/*.csv')

maxell_dataframe = pd.DataFrame()


for file in maxellFilesPath:
    print("Appending file"+file+" to dataframe")
    maxell_dataframe = maxell_dataframe.append(pd.read_csv(file,header=None,dtype={'Voltage':float,'WiFi TX count':int,'Last TX time': float, 'timestamp':np.datetime64, 'WiFi pin ON': int} ))

maxell_dataframe.columns=['Voltage','WiFi TX count','Last TX time','timestamp','WiFi pin ON']  
maxell_dataframe['timestamp']= maxell_dataframe['timestamp'].astype(np.datetime64)



### Data cleaning, voltage under 1.5V means no battery connected.
print("Cleaning invalid data")
maxell_dataframe = maxell_dataframe.drop(maxell_dataframe[maxell_dataframe.Voltage < 1.5].index)

### Data sorting
print("Checking data is sorted")
maxell_dataframe = maxell_dataframe.sort_values(by=['timestamp'])
#maxell_dataframe.set_index('timestamp', drop=True, append=False, inplace=True, verify_integrity=False)
#maxell_dataframe.sort_index()


### Time axis transformation, from timestamp to time delta and to hours
print("Processing time axis")
timeStartMaxell = maxell_dataframe.iloc[0,3]
timeAxisMaxell = ((maxell_dataframe.iloc[:,3] - timeStartMaxell))
timeAxisMaxell = (timeAxisMaxell / np.timedelta64(1, 's')) *1/3600


print(maxell_dataframe.describe())
print(maxell_dataframe.info())
```

 * Measurement data is loaded into data structures.
 * Data is cleaned from measurements before and after the battery is connected. 
 * Consistency checks are performed.
 * Time axis is tranformed from nanoseconds timestamps to hours

```
fig = plt.figure(figsize=(30,10))
plt.yticks(np.arange(0, 5, step=0.05))
plt.xticks(np.arange(0, 70, step=1))
plt.grid(True,which="both")
plt.xlabel('time (hours)')
plt.ylabel('voltage (V)')
plt.title("PKCell Ultra alkaline 3xAAA series Voltage")
plt.axhline(y=2.9,linewidth=2, color='r')
plt.plot(timeAxisPkcell, pkcell_dataframe.iloc[:,0],linestyle='', marker='o', markersize=0.5)

fig.savefig('/content/gdrive/My Drive/PROJECTS/B2B/PIKOLIN/SMM+/Energy estimation/Battery tests/PKCELL_full_discharge_curve.png', dpi = 1024)
plt.show()
```

### Graph comments:

This is the discharge voltage curve for the PKCell powered device. Two curves can be clearly observed, one for the device drawing the average of 50uA (upper curve) and another for the WiFi enabled status of the device were current consumption is x100 and hence voltage drops. 

The measurement script hang after 3 hours but the device is not stopped. Since measurements are done with absolute timestamps (a date with time) we can check after re-starting mesurement if it is consistent. Also the counting of TX events is done online in real cloud server (real data is transmitted and confirmed to be OK).

WiFi will continue working below the nominal working voltage of 3.3 trought 2.9 which is the least voltage it can work. Below 2.9V with the WiFi chip enabled, the device will keep rebooting because voltage will drop below the Nordic minimum working voltage. 

There is a horizontal line at 2.9V to clearly mark the voltage where it dies.

The upper graph (device without WiFi enabled) shows that with the WiFi disabled, the device battery life will extend until voltage drops below flash memory chip minimum voltage (2.3V).

```
fig = plt.figure(figsize=(30,10))
plt.grid(True,which="both")
plt.yticks(np.arange(0, 5, step=0.05))
plt.xticks(np.arange(0, 70, step=1))
plt.xlabel("time (hours) ")
plt.ylabel("voltage (V)")
plt.title("Maxell 3xAAA series Voltage")
plt.axhline(y=2.9,linewidth=2, color='r')
plt.plot(timeAxisMaxell, maxell_dataframe.iloc[:,0],linestyle='', marker='o', markersize=0.5, color = 'green')

fig.savefig('/content/gdrive/My Drive/PROJECTS/B2B/PIKOLIN/SMM+/Energy estimation/Battery tests/MAXELL_full_discharge_curve.png', dpi = 1024)
plt.show()
```
